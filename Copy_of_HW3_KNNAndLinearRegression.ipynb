## Part 2: Linear Regression

Train a linear regression model using the normal equation and verify that the results coincide with the results from `sklearn`.

We will use the advertisement revenue data used in Week 5 and build a linear regression model on `radio` and `sales`.

```python
url = "https://www.statlearning.com/s/Advertising.csv"
advertising = pd.read_csv(url, index_col=0)
advertising.head()
```

1. Split the data into 85% training data and 15% test data.

```python
raw_data = pd.read_csv(url) 
training_data, test_data = train_test_split(raw_data, test_size=0.15)
```

2. Build a linear regression model that predicts `sales` with the `radio` feature using `sklearn`. Display the model parameters and its test MSE.

```python
from sklearn.linear_model import LinearRegression
model_lr = LinearRegression()
model_lr.fit(advertising[['radio']], advertising[['sales']])

print(model_lr.coef_) # beta_1
print(model_lr.intercept_) # beta_0
```

```python
m = model_lr.coef_[0, 0]   # slope
b = model_lr.intercept_[0] # y-intercept

plt.plot(advertising['radio'], advertising['sales'], 'b.')
x_coordinates = np.array([0,60])
y_coordinates = x_coordinates * m + b
plt.plot(x_coordinates, y_coordinates, 'g-')

beta0 = 9.31
beta1 = 0.20
```

3. Calculate the parameter values with the normal equation. These values should be the same as the values shown in Step 2.

```python
def get_squared_error(beta0, beta1, advertising, i):

    xi = advertising.loc[i, 'radio']
    yi = advertising.loc[i, 'sales']
   

    
    f_xi = beta0 + beta1 * xi

   
    squared_error = (yi - f_xi) ** 2

    return squared_error
```

```python
list_errors = []
list_errors = [get_squared_error(beta0, beta1, advertising, ind) for ind in advertising.index]

MSE = sum(list_errors) / len(list_errors)

print("MSE:", MSE)
```
